```python
# prompt: products_df=pd.read_csv('https://wagon-public-datasets.s3.amazonaws.com/certification/da-ds-de/products.csv')

import pandas as pd
sales_df = pd.read_csv('https://wagon-public-datasets.s3.amazonaws.com/certification/da-ds-de/orders.csv')
products_df = pd.read_csv('https://wagon-public-datasets.s3.amazonaws.com/certification/da-ds-de/products.csv')
sales_df
```

```python
# prompt: sales_df.duplicated.sum()

sales_df.duplicated().sum()
sales_df=sales_df.drop_duplicates()
sales_df

products_df.duplicated().sum()
products_df=products_df.drop_duplicates()
products_df


```

```python
ventes_prods= prod_sales_df.groupby('ProductID')['NumberOfUnits'].sum().reset_index()
print(ventes_prods)
plt.figure(figsize=(10, 6))
sns.barplot(x = 'ProductID', y = 'NumberOfUnits', data = ventes_prods)
plt.show
```

```python
ventes_prod_tier = prod_sales_df.groupby('Tier')['NumberOfUnits'].sum().reset_index()
print(ventes_prod_tier)
plt.figure(figsize=(10, 6))
sns.barplot(x = 'Tier', y = 'NumberOfUnits', data = ventes_prod_tier)
plt.show
```
